```python
# Scrapes until canceled

i = 1
while True:
    sleep(random.randrange(1, 3))
    ts = tables(randowiki)
    try:
        vals = list()
        for k, v in ts[0].context.items():
            vals.append(str(v))
        print(f' * {vals}\n * {ts[0].record}\n\n')
        table = dict()
        table['i'] = i
        table['context_vals'] = vals
        table['table_record'] = ts[0].record[0]
        with open('Data/tableTest.jsonl', 'a') as tst:
            tst.write(f'{json.dumps(table)}\n')
            i += 1
    except IndexError:
        pass
    except TypeError:
        pass
```
